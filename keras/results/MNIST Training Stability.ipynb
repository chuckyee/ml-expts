```python
from __future__ import print_function, division

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def load_data(filename):
    with open(filename, "r") as f:
        data = f.readlines()
    data = [[float(x) for x in line.split()] for line in data]
    test_loss = data[-2][0] 
    test_accuracy = data[-1][0]
    return data[:-2], (test_loss, test_accuracy)
```

```python
prefix = "./vary-units/rmsprop"
suffixes = "32 64 128 200 256 300 400 512".split()
filenames = ["{}-{}".format(prefix, suffix) for suffix in suffixes]

for i,filename in enumerate(filenames):
    data, (test_loss, test_accuracy) = load_data(filename)
    plt.plot([x[0] for x in data],
             linewidth=2, alpha=0.8,
             color=plt.cm.viridis(i/(len(filenames)-1)),
             label=filename)
plt.legend()
plt.xlabel("Epoch")
plt.ylabel("Training loss")
plt.title("Varying hidden layer size (RMSprop)")
```

```python
prefix = "./vary-optimizer/"
optimizers = "sgd rmsprop adagrad adadelta adam adamax nadam".split()
suffix = "256"
filenames = ["{}{}-{}".format(prefix, optimizer, suffix) for optimizer in optimizers]

for filename in filenames:
    data, (test_loss, test_accuracy) = load_data(filename)
    plt.plot([x[0] for x in data], 'o-',
             linewidth=2, alpha=0.8,
             label=filename)
plt.legend()
plt.xlabel("Epoch")
plt.ylabel("Training loss")
plt.title("Varying optimizer (256 hidden units)")
```

```python
prefix = "./vary-learning-rate/rmsprop-256"
rates = "0.1 0.01 0.001 0.0001".split()
filenames = ["{}-{}".format(prefix, rate) for rate in rates]

for filename in filenames:
    data, (test_loss, test_accuracy) = load_data(filename)
    plt.plot([x[0] for x in data], 'o-',
             linewidth=2, alpha=0.8,
             label=filename)
plt.legend()
plt.xlabel("Epoch")
plt.ylabel("Training loss")
plt.title("Varying learning rate (RMSprop, 256 hidden units)")
```
